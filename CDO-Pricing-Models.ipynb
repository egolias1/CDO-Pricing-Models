```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random variables
np.random.seed(0)
N = 1000
X = np.random.normal(0, 1, N)
Y = np.random.normal(0, 1, N)

# Sort the variables
X_sorted = np.sort(X)
Y_sorted = np.sort(Y)

# Calculate the ranks
rank_X = np.argsort(np.argsort(X))
rank_Y = np.argsort(np.argsort(Y))

# Calculate the empirical copula
emp_copula = (rank_X + 1) / (N + 1)

# Plot the variables and the copula
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

# Scatter plot of the variables
axs[0].scatter(X_sorted, Y_sorted, color='b', alpha=0.5)
axs[0].set_xlabel('X')
axs[0].set_ylabel('Y')
axs[0].set_title('Scatter Plot of Variables')

# 2D plot of the copula
axs[1].scatter(emp_copula, Y_sorted, color='r', alpha=0.5)
axs[1].set_xlabel('Empirical Copula')
axs[1].set_ylabel('Y')
axs[1].set_title('2D Plot of Empirical Copula')

plt.tight_layout()
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph
G = nx.DiGraph()

# Add nodes for tranches and assets
G.add_node("CDO")
G.add_nodes_from(["Equity", "Mezzanine", "Senior"])
G.add_nodes_from(["Asset 1", "Asset 2", "Asset 3"])

# Add edges to represent the structure
G.add_edge("CDO", "Equity")
G.add_edge("CDO", "Mezzanine")
G.add_edge("CDO", "Senior")
G.add_edge("Asset 1", "Equity")
G.add_edge("Asset 2", "Mezzanine")
G.add_edge("Asset 3", "Senior")

# Set node positions for better visualization
pos = {
    "CDO": (0, 0),
    "Equity": (-1, 1),
    "Mezzanine": (0, 1),
    "Senior": (1, 1),
    "Asset 1": (-1, 2),
    "Asset 2": (0, 2),
    "Asset 3": (1, 2),
}

# Draw the graph
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_size=1000, node_color="lightblue",
        arrowsize=20, font_size=12, font_weight="bold", edge_color="gray")
plt.title("Structure of a Collateralized Debt Obligation (CDO)")
plt.axis("off")
plt.show()
```
